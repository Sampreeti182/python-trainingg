```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("CustomerData").getOrCreate()

# Create DataFrame
columns = ["customer_id", "name", "city", "age", "annual_spend", "category"]
df = spark.createDataFrame(data, columns)

df.show(10)
```

```python
df.select("city").distinct().show()
```

```python
df.select("customer_id","name","annual_spend").show()
```

```python
df.filter(df.annual_spend>60000).show()
```

```python
df.filter((df.city=="Delhi")&(df.age<30)).show()
```

```python
from pyspark.sql.functions import col
df=df.withColumn("spend_lakh",col("annual_spend")/100000)
df.show()
```

```python
from pyspark.sql.functions import when
df=df.withColumn("customer_type",when (df.annual_spend>70000,"Premium").otherwise("Standard"))
df.show()
```

```python
df.filter(df.name.startswith("A")).show()
```

```python
df.filter(df.category.isin("Clothing","Electronics")).show()
```

```python
from pyspark.sql.functions import upper
df=df.withColumn("city_upper",upper(df.city))
df.show(

)
```

```python
from pyspark.sql.functions import avg
avg_age=df.select(avg("age")).first()[0]
df.filter(df.age<avg_age).show()
```

```python
from pyspark.sql.functions import substring
df=df.withColumn("first_letter",substring("name",1,1))
df.show()
```

```python
from pyspark.sql.functions import concat,lit,substring
df=df.withColumn("masked_id",concat(lit("***"),substring("customer_id",-2,2)))
df.show()
```
